```python
import pandas as pd
import ast
```

```python
allData = pd.read_csv('data/CombinedData.csv')
allData
```

```python
data = allData[allData['QType']=='continuous']
data
```

```python
maxes = []
mins = []
for choice in list(data['AnswerChoices']):
    choice = ast.literal_eval(choice)
    if 'scale' in choice:
        maxes.append(choice['scale']['max'])
        mins.append(choice['scale']['min'])
    else:
        maxes.append(choice['max'])
        mins.append(choice['min'])
```
