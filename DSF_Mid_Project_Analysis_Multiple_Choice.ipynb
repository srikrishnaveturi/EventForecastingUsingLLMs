```python
import pandas as pd
predictions = pd.read_csv('data/Multiple/WizardMulti0-925.csv')
predictions
```

```python
Ques_Multiple = pd.read_csv('data/Multiple/CombinedMultipleDataWithCategories.csv')
Ques_Multiple
```

```python
MultipleChoice = pd.merge(Ques_Multiple, predictions[['predictions']], left_index=True, right_index=True)
MultipleChoice
```

```python
import ast

def find_matching_string(answer_choices, predictions_text):
    predictions_text_set = set(predictions_text.lower().split())

    for i, choice in enumerate(answer_choices, start=1):
        choice_words_set = set(choice.lower().split())
        intersection = choice_words_set.intersection(predictions_text_set)

        if len(intersection) >= 3:
            return choice, i

    return None, None


for i in range(len(MultipleChoice)):
    answer_choices_str = MultipleChoice['AnswerChoices'][i]
    predictions_text = str(MultipleChoice['predictions'][i])


    if pd.notna(answer_choices_str):
        # Convert string representation of list to an actual list
        answer_choices = ast.literal_eval(answer_choices_str)

        # Apply the function to create new 'MatchingString' and 'MatchingIndex' columns
        matching_string, matching_index = find_matching_string(answer_choices, predictions_text)


        MultipleChoice.at[i, 'MatchingString'] = matching_string
        MultipleChoice.at[i, 'MatchingIndex'] = matching_index


MultipleChoice[['predictions', 'AnswerChoices', 'MatchingString']]
```
