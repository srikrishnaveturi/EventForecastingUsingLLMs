```python
finalPredictions = []
finalPredictions.extend(list(data1['predictions']))
finalPredictions.extend(list(data2['predictions']))
finalPredictions.extend(list(data3['predictions']))
finalPredictions.extend(list(data4['predictions']))
finalPredictions.extend(list(data5['predictions']))
finalPredictions.extend(list(data6['predictions']))
finalPredictions.extend(list(data7['predictions']))
finalPredictions.extend(list(data8['predictions']))
len(finalPredictions)
```

```python
actualData = pd.read_csv('data/CombinedBinaryData.csv')
actualData
```

```python
def get_value(text):
    if 'yes' in text.lower():
        return 1
    elif 'no' in text.lower():
        return 0
    else:
        return None
```

```python
y_true = []
y_pred = []
noneCount = 0
for i in range(len(finalPredictions)):
    temp1 = get_value(finalPredictions[i])
    if temp1 is None:
        noneCount += 1
        continue
    y_pred.append(temp1)
    y_true.append(get_value(actualData['Answer'][i]))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_true, y_pred)
```

```python
y_pred.count(1)
```

## Analysis with respect to resolved year

```python
actualData = actualData[['Year', 'Answer']]
actualData['Prediction'] = finalPredictions
actualData
```

```python
y_true_before = []
y_pred_before = []
y_true_after = []
y_pred_after = []
noneCount = 0

for i in range(len(finalPredictions)):
    temp1 = get_value(finalPredictions[i])
    if temp1 is None:
        noneCount += 1
        continue
    if int(actualData['Year'][i]) > 2022:
        y_pred_after.append(temp1)
        y_true_after.append(get_value(actualData['Answer'][i]))
    else:
        y_pred_before.append(temp1)
        y_true_before.append(get_value(actualData['Answer'][i]))        
```

```python
len(y_pred_before)
```

```python
confusion_matrix(y_true_before, y_pred_before)
```

```python
confusion_matrix(y_true_after, y_pred_after)
```
