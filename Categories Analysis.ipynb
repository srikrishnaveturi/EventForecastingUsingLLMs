```python
import pandas as pd
from tqdm import tqdm
import requests
```

```python
finalData = pd.DataFrame({
    'question': [],
    'qtype': [],
    'answer': [],
    'resolve_time': [],
    'active_state': []
})
nxt = 'https://www.metaculus.com/api2/questions/?status=RESOLVED&limit=100'
for counter in tqdm(range(100)):
    response = requests.get(nxt)
    nxt = response.json()['next']
    for res1 in response.json()['results']:
        if res1['active_state'] == 'RESOLVED':
          finalData.loc[len(finalData.index)] = [res1['title'], res1['possibilities'], res1['resolution'], res1['resolve_time'],res1['active_state']]
    if nxt is None:
        break
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
# Assuming you have the 'dat' DataFrame with the attributes 'QuestionText', 'Tags', and 'Description'

# Define the categories
categories = {
    1: 'Technology',
    2: 'Political',
    3: 'Sports',
    4: 'Health and Epidemiology',
    5: 'Stock Market',
    6: 'Supply Chain'
}
```

```python
# Concatenate text columns into a single column
dat.Description = dat.Description.fillna("")
dat['CombinedText'] = dat['QuestionText'] + ' ' + dat['Tags'] + ' ' + dat['Description']

# Load a pre-trained DistilBERT model for sentence embeddings
model = SentenceTransformer('distilbert-base-nli-mean-tokens')

# Encode category texts and store them in a dictionary
category_embeddings = {}
for category_id, category_text in categories.items():
    embeddings = model.encode(category_text, convert_to_tensor=True)
    category_embeddings[category_id] = embeddings

def encode_with_check(x):
    if pd.isna(x):
        return None
    return model.encode(x, convert_to_tensor=True)
```

```python
# Apply DistilBERT embeddings to the combined text
dat['Embeddings'] = dat['CombinedText'].apply(encode_with_check)
```

```python
temp = [x.cpu() for x in dat['Embeddings']]
dat['Embeddings'] = temp
```

```python
# Extract embeddings for each category and compute cosine similarity
for category_id, category_text in categories.items():
    category_embedding = category_embeddings[category_id].cpu()
    dat[category_text + '_Similarity'] = dat['Embeddings'].apply(lambda emb: np.dot(emb, category_embedding) / (np.linalg.norm(emb) * np.linalg.norm(category_embedding)))

# Display the resulting DataFrame with category similarities
dat[['QuestionText', 'QType', 'Answer', 'AnswerChoices', 'Year', 'Tags', 'Description'] + [category_text + '_Similarity' for category_text in categories.values()]]
```

```python
similarity_columns = [category_text + '_Similarity' for category_text in categories.values()]

# Function to get the category with the highest similarity for each row
def get_similar_categories(row):
    similarity_cats = [col.split('_')[0] for col in similarity_columns if row[col] > 0.15]
    return similarity_cats

dat['Categories'] = dat.apply(get_similar_categories, axis=1)

dat[['QuestionText', 'QType', 'Answer', 'AnswerChoices', 'Year', 'Tags', 'Description', 'Categories']]
```
